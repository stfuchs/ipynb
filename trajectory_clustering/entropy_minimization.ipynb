```python
%matplotlib inline
import numpy as np
import numpy.random as rnd
```

```python
def normalize(x):
    """normalize discrete probability distributions
    
    x may be a 2D array ([n,m]) with n distributions of size m
    or 1D array ([m]) with a single distribution of size m"""
    if(len(x.shape)==2): return (x.T/np.sum(x,axis=1)).T
    else: return (x/np.sum(x))

def entropy(p):
    """compute entropy of a discrete probability distribution"""
    return - np.sum(p*np.nan_to_num(np.log(p)))
```

```python
def measure(F, noise=1., shuffle=True):
    """takes the true labels F, shuffles them around 
    and adds some uniform noise"""
    n,k = F.shape
    err = noise*rnd.rand(n,k)
    idx = np.array(range(k))
    idx_true = np.array(range(k))
    if shuffle:
        rnd.shuffle(idx)
        idx_true = list(idx_true[np.argsort(idx)])
    idx = list(idx)
    Z = F[:,idx]
    if shuffle: print("Z -> F: %s \tF -> Z: %s"%(idx_true,idx))
    return normalize(Z+err),idx_true
```

Given a set $\mathcal{D}^{(t)} = \{x_i^{(t)}\}_{i=1}^n$ of $n$ trajectories $x_i$ at timestep $t$ and their affinity
matrix $\mathbf{A}^{(t)} \in [0,1]^{n\times n}$ with $a_{ij} = k(x_i^{(t)},x_j^{(t)})$
being the result of some kernel function.

In the following let 

$$\mathbf{F} \in \{0,1\}^{n \times k}$$

be the true matrix that assigns each trajectory to exactly one object (or cluster) $c \in \{1,...,k\}$. Hence

$$\sum_{j=1}^k \mathbf{F}_{ij} = 1 \qquad \forall i \in \{1,...,n\} .$$

Now suppose in every time step $t$ we can determine which trajectories belong together by 
fitting a Gaussian mixture model (GMM) to the current affinity matrix $\mathbf{A}^{(t)}$.
This basically assignes a label $l^{(t)} \in \{1,...,k\}$ to every trajectory $x_i^{(t)}$ and provides
us with a discrete probability distribution $P(l^{(t)}|x_i^{(t)}, \mathcal{D}^{(t)})$ for every trajectory
over all labels $l^{(t)}$.
The problem is that the labels of GMM do not necessarily correspond to the same cluster $c$ 
and change with every new fitting ($l^{(t)} \neq l^{(t+1)}$).
This means we need to find a mapping $w^{(t)}: l^{(t)} \mapsto c$ after each fitting.

Lets start by assuming that the labels are always correct, so that $l^{(t)} := l^{(t+1)} := c$.
(I think this notation is wrong. Actually the space of $l$ and $c$ are the same, meaning that $l=1$ and $c=1$ both
refer to the same object. Of course $l$ and $c$ don't have to be the same value since trajectories can be assigned
to different objects when new data is available.)
We still need to propagate the measured uncertainty over time. We do this by applying a Bayes filter

$$ \begin{align}
    P(c^{(t)}|z_i^{(t)},x_i^{(t)}) &= \eta P(z_i^{(t)}|c^{(t)},x_i^{(t)})
        \sum_{c^{(t-1)}} P(c^{(t)}|c^{(t-1)}) P(c^{(t-1)}|z_i^{(t-1)},x_i^{(t)})\\
        &=\eta P(z_i^{(t)}|c^{(t)},x_i^{(t)}) P(c^{(t)}|z_i^{(t-1)},x_i^{(t)})
\end{align}$$

for every trajectory $x_i$. If we know that $l:=c$ we could just replace the measurement uncertainty $P(z_i^{(t)}|c^{(t)},x_i^{(t)})$ with the results of the GMM fit so that

$$ 
    P(z_i^{(t)}|c^{(t)},x_i^{(t)}) = P(l^{(t)}|c^{(t)},x_i^{(t)}) := P(l^{(t)}|x_i^{(t)}, \mathcal{D}^{(t)}).
$$

But since we have $w(l):=c$ we need to look for

$$
    P(w(l^{(t)})|c^{(t)},x_i^{(t)}) = P(z_i^{(t)}|c^{(t)},x_i^{(t)};\omega^{(t)})
$$

where $\omega^{(t)} \in \{1,...,k!\}$ parametrizes all possible mappings between space of $l$ and space of $c$.
We could choose $\omega$ such that the resulting probability distribution $P(c^{(t)}|z_i^{(t)},x_i^{(t)})$ for all
trajectories $x_i$ becomes peaked for a single object class. In other words we want to minimize the entropy

$$ H(x) = - \sum_i p(x_i)\log p(x_i) $$
or in our case
$$
H(c^{(t)},x_i^{(t)},\omega^{(t)}) =
    - \sum_j P_j(c^{(t)}|z_i^{(t)},x_i^{(t)};\omega^{(t)}) \log P_j(c^{(t)}|z_i^{(t)},x_i^{(t)};\omega^{(t)})
$$

and choose $\omega$ such that

$$ \underset{\omega^{(t)}}{\text{min}} \qquad \sum_i H(c^{(t)},x_i^{(t)},\omega^{(t)}). $$

However with increasing number of clusters $k$ this quickly becomes intractable.
Therefore we need to approach this problem more heuristically. The first assumption is that the number of trajectories that
switch to a different cluster is rather small. This already is quite a strong restriction but assuming that splitting an
existing cluster (where a large number of trajectories get reassigned) usually also results in an increase of the total number of clusters and should make up for it.

The other assumption is that in order to minimize the entropy, one condition is to combine the maximal probabilites of both measurement and prediction. In other words, high correlation between the probabilities is a strong indicator to maximizes
the resulting probability. Hence we create a single correlation matrix $\mathbf{M} \in \mathbb{R}^{k \times k}$ with entries

$$ \mathbf{M}_{pq} = \sum_i^n P_p(c^{(t)}|z_i^{(t-1)},x_i^{(t)}) P_q(z_i^{(t)}|c^{(t)},x_i^{(t)}) $$

for all trajectories $x_i$, where $P_p( ... )$ and $P_q( ... )$ denote the p'th and q'th probability of the corresponding
distribution. From $\mathbf{M}$ we $k$-times pick the indices $p,q$ with the maximum value and set row and column $\mathbf{M}_{p,1:k} = \mathbf{M}_{1:k,q} = 0$ to prevent from picking there again. Every pick denotes a mapping $l_p \mapsto c_q$.

```python
def predict(X,prob):
    """apply transition probability"""
    n,k=X.shape
    p_t = np.ones([k,k])*(1.-prob)/(k-1.)
    p_t[np.diag_indices_from(p_t)]=prob
    for i in range(n):
        X[i,:] = p_t.dot(X[i,:])
    return X

def maxmax(M):
    """pick indices from correlation matrix"""
    n,k = M.shape
    idx = [0]*k
    for ki in range(k):
        i,j = np.unravel_index(np.argmax(M),M.shape)
        M[i,:] = 0
        M[:,j] = 0
        idx[i]=j
    return idx

def indices(X,Z):
    n,k=X.shape
    corr = np.zeros([k,k])
    for x,z in zip(X,Z):
        corr += np.mat(x).T*np.mat(z)
    idx = maxmax(corr)
    print("Z -> X: %s (max correlation)"%idx)
    return idx

def update(X,Z):
    """do the same as maxmax() and indices(), 
    but create a linear transform M based on the correlation matrix"""
    n,k=X.shape
    corr = np.zeros([k,k])
    for x,z in zip(X,Z):
        corr += np.mat(x).T*np.mat(z)
    M = np.zeros([k,k])
    for ki in range(k):
        i,j = np.unravel_index(np.argmax(corr),corr.shape)
        corr[i,:]=0
        corr[:,j]=0
        M[i,j]=1.
    for ni in range(n):
        X[ni,:] = X[ni,:]*np.array(M.dot(Z[ni,:]))
    return normalize(X)

def bruteforce(X,Z):
    """only works for k=3,
    computes the entropy for every combination and returns the minimum"""
    n,k=X.shape
    Ah=np.array([[(0, 0),(0, 1),(0, 2),(0, 0),(0, 1),(0, 2)],
                 [(1, 1),(1, 2),(1, 0),(1, 2),(1, 0),(1, 1)],
                 [(2, 2),(2, 0),(2, 1),(2, 1),(2, 2),(2, 0)]])
    A = (Ah[:,:,0]*k+Ah[:,:,1]).T
    H = np.zeros([6,1])
    for x,z in zip(X,Z):
        cov = np.mat(x).T*np.mat(z)
        zx = np.zeros([6,3])
        for i in range(6):
            zx[i,:] = cov[np.unravel_index(A[i,:],cov.shape)]
            H[i] += entropy(normalize(zx[i,:]))
    
    w = np.argmin(H)
    idx = list(Ah[:,w,1])
    print("Z -> X: %s (brute force)"%idx)
    print("H(x,w) = %s"%H.T)
    return idx
```

```python
np.set_printoptions(precision=3,suppress=True)
K = 3 # number of clusters
N = 10 # number of features
measurement_noise = 1.
transition_likelihood = .6

F=np.zeros([N,K])
for i,v in enumerate(rnd.randint(0,K,N)): F[i,v]=1
print("true labels F:")
print(F.T)

X,idx_true = measure(F,measurement_noise,False)
print("\ninitial distribution:")
print(X.T)
print("")

for i in range(10):
    Z,idx_true = measure(F,measurement_noise)
    X = predict(X,transition_likelihood)
    #X = update(X,Z)
    idx = indices(X,Z)
    if K==3: bruteforce(X,Z)
    X = normalize(X*Z[:,idx])
    print(X.T)
    print("")

print("")
print("distribution of X:")
print(X.T)
print("")
```
