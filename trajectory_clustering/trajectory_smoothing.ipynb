```python
%matplotlib inline
from IPython.html.widgets import interactive
from IPython.html import widgets
from IPython.display import display, Image
import glob

from numpy import *
```

```python
def selectDataset(dataset):
    fnames = glob.glob(dataset+'/*.txt')
    img = glob.glob(dataset+'/*.jpg')
    print "Preview:"
    if len(img)!=0 : display(Image(filename=img[0], width=320))
    else: print "   --- not available --- "
    print "Found {0} files".format(len(fnames))
    return fnames
data_select = interactive(selectDataset, dataset = widgets.RadioButtonsWidget(values=glob.glob('data/*')))
display(data_select)
```

```python
class Trajectory:
    def __init__(self, id_):
        self.id_ = id_
        self.t_prev = 0.
        self.z = []

    def update_self(self, z, t):
        self.z.append(mat(z).T)
        self.t_prev = t

    def qt(self): return self.x[-1][:3]
    def dqt(self): return self.x[-1][3:]
    def Jg(self,dt):
        J = mat(identity(6))
        J[0,4] = J[1,4] = J[2,5] = dt
        return J
    def Jh(self):
        return mat(hstack([identity(3), zeros([3,3])]))

    def update_first(self, z2, t2):
        z2 = mat(z2).T
        if linalg.norm(z2 - self.z[-1]) > .3: return
        self.x = [ vstack( [z2, (z2-self.z[-1])/(t2-self.t_prev)] ) ]
        self.Cx = mat(identity(6)*.1)
        self.z.append(z2)
        self.t_prev = t2

    def update_smoothed(self, z, t):
        z = mat(z).T
        dt = (t - self.t_prev)
        xt = vstack( [self.qt() + dt * self.dqt(), self.dqt()] )
        Cx = self.Jg(dt) * self.Cx * self.Jg(dt).T + identity(6)*.001
        diff = vstack([z, (z - self.z[-1])/dt]) - xt
        if len(self.z) < 5:
            if linalg.norm(self.z[-1] - z) > .3: return
        elif (linalg.norm(diff)) > .15:
            return
        #K = Cx * self.Jh().T * linalg.inv(Cx[:3,:3]+identity(3)*.1)
        K = Cx * linalg.inv(Cx+identity(6)*.1)
        self.x.append( xt + K*diff )
        self.Cx = (identity(6) - K)*Cx
        self.z.append(z)
        self.t_prev = t
```

```python
T1 = []
T2 = []
Z = []

id_ = 0
for i in range(len(data_select.result)):
    x = loadtxt(data_select.result[i])
    if x.shape > 1:
        T1.append(Trajectory(id_))
        T2.append(Trajectory(id_))
        for j in range(len(x)): Z.append( (x[j,-1], id_, x[j,:-1]) )
        id_ += 1
Z.sort()

for z in Z:
    time, traj_id, point = z
    T1[traj_id].update_self(point,time)
for z in Z:
    time, traj_id, point = z
    if len(T2[traj_id].z) == 0:
        T2[traj_id].update_self(point,time)
    elif len(T2[traj_id].z) == 1:
        T2[traj_id].update_first(point,time)
    else:
        T2[traj_id].update_smoothed(point,time)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig,ax = plt.subplots(nrows=2, ncols=1, figsize=(14,28), subplot_kw=dict(projection='3d'))
for axi in ax:
    axi.grid()
    axi.set_xlabel('z')
    axi.set_ylabel('x')
    axi.set_zlabel('y')
    axi.set_xlim([-1.8, -.6]) #z
    axi.set_ylim([- .5,  .5]) #x
    axi.set_zlim([- .4,  .2]) #y

for t in T1:
    tmp = array(hstack(t.z))
    ax[0].plot(-tmp[2,:],tmp[0,:],-tmp[1,:])
for t in T2:
    tmp = array(hstack(t.z))
    ax[1].plot(-tmp[2,:],tmp[0,:],-tmp[1,:])
```
