```python
%matplotlib inline
from IPython.html.widgets import interactive, interact
import glob
from numpy import *
import matplotlib.pyplot as plt
import re
```

```python
labels = load('labels/labels.npy')
tmp_ids = load('labels/ids.npy')
tmp_ids = tmp_ids >> 32
ids = {}
for i in range(len(tmp_ids)):
    ids[tmp_ids[i]] = i
stamps = load('labels/timestamps.npy')
fnames = glob.glob('labels/traj/*.npy')
knames = glob.glob('labels/kernels/*.npy')
inames = glob.glob('labels/ids/*.npy')
kernels = [array([])]*len(knames)
kernel_ids = [array([])]*len(inames)
for k in knames:
    i = int(re.search("(\d+).npy",k).group(1))
    kernels[i] = load(k)
for k in inames:
    i = int(re.search("(\d+).npy",k).group(1))
    kernel_ids[i] = load(k) >> 32
```

```python
def show_sorted(index):
    indices = argsort(labels[:,index])
    fig, ax = plt.subplots(figsize=(12,16))
    im = ax.imshow(labels[indices], interpolation='none')
    ret = ax.set_xticks([index])
    ret = ax.set_yticks(range(len(tmp_ids)))
    yticks = ax.set_yticklabels(tmp_ids[indices],fontsize=8)
interactive(show_sorted, index=(0,labels.shape[1]-1,1))
```

```python
x = []
tmp_name_ids = []
maxlen = 0
for fname in fnames:
    x.append(load(fname))
    tmp_name_ids.append(int(re.search("(\d+).npy",fname).group(1)))
for xi in x:
    maxlen = max(maxlen, xi.shape[0])
T = []
name_ids = []
for i in range(len(x)):
    if x[i].shape[0]<maxlen: continue
    T.append(x[i])
    name_ids.append(tmp_name_ids[i])

T_stamps = []
for t in T:
    T_stamps.append(t[:,-1])
T_stamps = nanmax(vstack(T_stamps),axis=0)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

colors = ['b','g','r','c','m','y','k']

ids
def show_lines(index):
    T_index = 0
    for i in range(len(T_stamps)):
        if T_stamps[i] > stamps[index]:
            T_index = i
            break
    print("Selected timestamp is {}".format(stamps[index]-stamps[0]))
    fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(18,12) ,subplot_kw=dict(projection='3d'))
    ax.grid()
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_xlim([-.6,.5])
    ax.set_ylim([-.7,.5])
    ax.set_zlim([0.5,1.5])
    #ax.view_init(elev=110,azim=-90)
    ax.view_init(elev=20,azim=-90)

    for i in range(len(T)):
        if name_ids[i] in ids:
            l = labels[ ids[ name_ids[i] ], index ]
            if l == -1: continue
            ax.plot(T[i][:T_index,0],T[i][:T_index,1],T[i][:T_index,2],color=colors[l])

interactive(show_lines,index=(0,labels.shape[1]-1,1))
```

```python
def show_kernel(index):
    fig, ax = plt.subplots(figsize=(18,18))
    K = kernels[index-1]
    k_ids = kernel_ids[index-1]
    l = labels[:,index]
    l = l[l!=-1]
    l_ids = tmp_ids[l!=-1]
    
    p_idx = argsort(l)
    Ks = K[p_idx,:]
    Ks = Ks[:,p_idx]
    im = ax.imshow(Ks, interpolation='none', vmin=.5, cmap=plt.get_cmap('Oranges'))
    #ticks = p_idx
    #ax.set_yticks(range(len(p_idx)))
    #yticks = ax.set_yticklabels(ticks,fontsize=8)
    #for i in p_idx):
    #    yticks[i].set_color(colors[ l[] ])
interactive(show_kernel, index=(0,len(kernels),1))
```

```python
def show_eigenvalues(index):
    val,vec=linalg.eig(kernels[index])
    idx = argsort(abs(val))
    fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(18,8))
    ax.set_ylim([0,20])
    ax.set_xlim([0,140])
    ax.grid()
    ax.plot(abs(val[idx]), 'x-')
interactive(show_eigenvalues,index=(0,len(kernels)-1,1))
```

```python
def plot_id(Id1, Id2):
    fig,ax = plt.subplots(nrows=3, ncols=1, figsize=(12,15))
    titles = ['position x', 'position y', 'position z']
    ax[0].set_ylim([- 1.,  1.]) #x
    ax[1].set_ylim([- .6,  .6]) #y
    ax[2].set_ylim([.5, 2.]) #z
    for i in range(len(ax)):
        ax[i].plot(T[Id1][:,3]-T_stamps[0],T[Id1][:,i],'x-')
        ax[i].plot(T[Id2][:,3]-T_stamps[0],T[Id2][:,i],'x-')
        ax[i].grid()
        ax[i].set_title(titles[i])
    
interactive(plot_id, Id1=(0,len(T)-1,1), Id2=(0,len(T)-1,1))
```
