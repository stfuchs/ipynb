#Inferring Object Hypotheses based on Feature Motion of Different Sources
##Introduction
Identifying objects of interest is a key requirement for every robot.
While navigating through an uncontrolled environment a robot has to avoid
potential obstacles such as pedestrians or cars on the road. Localization
and mapping tasks (SLAM) require to track landmarks in order to estimate
the relative sensor position. For manipulation one has to know about what
and how objects can be manipulated.

For all these tasks motion is either the mandatory element or
they can benefit from it. For example, showing the robot that a drawer
can be opened or a mug can be moved from the table yields valuable
information for many manipulation tasks. In fact the importance of motion for perception
systems has been subject to lots of research and originates from the
*Principles of Gestalt Psychology* by Max Wertheimer, Wolfgang Kohler
and Kurt Koffka \cite{Koffka_1935}. In particular the *common fate* principle
states that elements are perceived as grouped together, if they move
along a common and smooth path. This principle has been applied to various
computer vision problems such as image segmentation \cite{Ochs_Brox_2011},
\cite{Kenney_Buckley_Brock_2009}, object recognition (CITE) or estimation
of kinematic models \cite{Martin_Brock_2014} .

In order to exploit the information from motion one has to address certain
difficulties. Typically measurements not only consists of data about
the moving objects, but also of the image background or clutter measurements.
Therefore dealing with sensor noise and background separation is an essential
requirement.

Having the luxury to know the number of objects a priori is rarely the case
in general tracking situations. So one not only has to estimate the number
of objects in the scene but also handle cases when objects enter or leave
the scene (usually refered to as object **birth** and **death**)
or when objects are occluded.

Dealing with multiple objects also requires to solve the data association
or correspondence problem. This means to associate each measurement to a
certain object. Early methods of multiple object tracking, when sensors
like radar or laser scanner did not provide high resolutions, it was possible
to resrict the association problem to one measurment per object. With increased 
resolution of modern sensor technology however,
this is not always applicable and leads to the
relaxed definition of **extended object** tracking.
This means that objects are not treated as point sources anymore but instead
feature a certain spatial extension that may cause multiple measurements per
object.


The goal of this thesis is to develop a general perception system that allows
to identify **entities** of interest based on observed motion from various
channels. These channels may consist of rgb image sequences obtained from
an ordinary camera feed, depth information from stereo reconstruction or
laser scanners, or even end effector trajectories such as the tip of a robot
arm that manipulates an object. The restriction we set to these channels is that
they allow us to extract motion in terms of point trajectories over time. Since
it can not be assured that measurement updates arise with a common frequency for
every channel, we define a feature trajectory $f_i$ as
$$
  f_i = \left\{ \left( \vect{z}_j^T, t_j \right) \right\}_{j=1}^{M_i}
$$
where $\vect{z}_j \in \mathbb{R}^N$ is the j-th observation of a feature in
the channel's space at time $t_j \in \mathbb{R}$, where $t_o<t_1< ... <t_{M_i}$.
Given this motion we now want to create hypotheses about trajectory combinations
that describe the path of the same entity.

**Estimation** is the process of inferring the value of a quantity of interest
from indirect, inaccurate and uncertain observations, the process of seclecting
a point from a continous space. **Decision** can be viewed as the *selection of one out 
of a set of discrete alternatives* -- the best choice. **Classification** of objects is a decision.
One can talk about estimation in a discrete-valued case with the
possibility of not making a choice but obtaining a some conditional probabilities of the various alternatives. This information can be used without making hard decisions. ** Tracking ** is the estimation of the state of a
moving object. **Filtering** is the estimation of the current state of a dynamic system from noisy data.
This amounts to "filtering out" the noise. **Smoothing** is the estimation of the states of as system at a time prior
to the current time.

A **track** is a state trajectory estimated from a set of measurements that have been associated with the same target.
These **data association** problems may be categorized according to what is associated with what:
* *measurement to measurement association* for track formation
* *measurement to track association* for track maintenance or updating
* *track to track association* for track fusion (as in multi-sensor systems)

##Challenges
Using trajectories is a simplification of the multiple extended object tracking
problem as such that solving the measurement correspondences and therefore the
motion in each channel is performed independently of the object estimation
process. Although there exist methods such as PHD-filter (CITE) that address
the problem in a more integrated manner, this approach quickly becomes
intractable when dealing with large amounts of measurements as they result from
image sequences. The remaining major challenges can be summarized as:

* at a certain time interval multiple trajectories can share the same path,
  although they actually originate from different objects. *Example:*
  someone puts a mug on a table. At first the person's hand and the mug are
  attached to each other and move along the same path.
  When the mug is released, it should become clear
  that the trajectories did not originate from a single object.
* trajectories may suddenly stop to receive updates, although the causing
  object can still be within the observation space (eg. FOV of the camera).
  This means the object could be partially or fully occluded and might
  reappear at one point later.
  Another possibility is that the object indeed left the observation area
  which can be treated as an object *death*. We
  should address both cases separatly.
  This however requires to be able to recover a lost trajectory.
  I'm not sure if this should be part of the motion estimation (tracker) or the
  object estimation level.
* as trajectories disappear, they may as well spawn regularly. This can
  happen due to new objects entering the observation space (e.g. object *birth*)
  or new parts of existing objects become visible.
* different channels do not run with the same frequency. Hence trajectories
  get updated asychronously.
* trajectory may become erroneous as such that they suddenly describe
  the path of a different object as they did earlier. This states the problem that
  the measurement correspondence required to be solved for each channel to obtain
  a point trajectory may not allways be correct.

## Ideas
We have currently looked at three different approaches.

### Gaussian Mixture Probability Hypothesis Density ({GM-PHD})
A brief summary: Formulating measurements and objects as random finite sets (RFS)
$$
  \mathrm{X}_k = \left\{x_k^{(i)}\right\}_{i=1}^{N_{x,k}}
$$
$$
  \mathrm{Z}_k = \left\{z_k^{(j)}\right\}_{j=1}^{N_{z,k}}
$$
at time step $k$. The {PHD}-filter propagates the first order statistical moment
of the probability density function $P(X)$
over time by defining the intensity function $v(x)$ as
$$
  \int_S v(x) \diff x = \int |\mathrm{X}\cap\mathrm{S}| P(\mathrm{X}) \diff \mathrm{X}.
$$
Hence the intergral of the intensity $v(x)$ over a defined region $\mathrm{S}$
(eg the observation area) yields the expected number of objects in this area.
In case of {GM-PHD} this function is represented as a Gaussian mixture
$$
  v_k(x) = \sum_{i=1}^{I_k} w_k^{(i)} N(x|m_k^{(i)},P_k^{(i)})
$$
where each component represents an object hypothesis. With every correction step
a large number of hypotheses are added as new mixture components. In theory
these components represent every possible object-measurement combination that
has been exisited until the current time step. It is clear that it would cause
the number of components to explode very quickly and some prune and merge
heuristics are required to reduce this number after every correction step.

> way to formalize the problem. However it is not clear this method can get anywhere
> close to an online realization (literature doesn't talk much about performance).

### Comparing Distance Variance of Trajectory-Pairs
With the assumption that spatial distance of two features of the same object is
constant under rigid transformation, we can pairwise compare trajectories and use
the variance of the distance measure as an indicator
whether they describe the same entity. Since trajectory updates arrive
asynchronously, we have to predict the position of all other trajectories based
on their previous motion when comparing with the updated one.

Yet unsolved is the way of how to cluster the trajectories. Example:
if the distance variance between trajectory $a$ and $b$ is small and also for $b$ and $c$,
this does not necessarly mean it has to be small for $a$ and $c$.


### Multi-Level Recursive State Estimation
This approach would use two different layers of recursive state estimation.
While the lower level deals with estimation of the feature locations, the
higher level estimates the object states. Although both layers are separated,
they are designed to communicate with each other about their updated states.
The feature locations on the first level are used for measurement update step
on the object level. In return, the updated object state used to predict the
possible future locations of the features to enhance the results of the tracker.

Assuming we already have a set of objects with assigned feature trajectories, 
the association of features to objects works in the following order:

* predict the new state of the object and the feature locations assigned to it,
* compare the feature prediction with their most recent measurement and
  unassign all features that could not be predicted correctly,
* assign free features to object that best predicted their location,
* correct the predicted object state by using the updated feature
  association and their last acquired measurement,
* try to create new objects from the remaining free features.
