# Problem 3: Quality of Taylor approximation

Consider the function
$$
\newcommand{\vx}{\mathbf{x}}
f(\vx) = \log\left( e^{\|\vx\|} + e^{-\|\vx\|} \right)
$$
with $\vx \in \mathbb{R}^n$.

```python
def f(x):
    h = float(sqrt(x.T*x))
    return log(exp(h) + exp(-h))
```

```python
X = arange(-2.,2.,0.05)
Y = [ f(x) for x in X ]

fig, axes = plt.subplots(figsize=(8,5))
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel(r'$f(x)$')
axes.plot(X,Y)
```



##### Gradient
$$ \nabla f = \frac{\vx}{\|\vx\|} \left( \frac{e^{\|\vx\|} - e^{-\|\vx\|}}{e^{\|\vx\|} + e^{-\|\vx\|}} \right) $$

```python
def df(x):
    h = sqrt(x.T*x)
    return x/h * (exp(h) - exp(-h))/(exp(h) + exp(-h))
```

```python
X = arange(-2.,2.,0.05)
Y = [ df(x) for x in X ]

fig, axes = plt.subplots(figsize=(8,5))
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel(r'$\nabla f(x)$')
axes.plot(X,Y)
```



##### Hessian
$$ \nabla^2 f = \frac{1}{\|\vx\|}
\left(
\alpha \mathbf{I} - \hat\vx \hat\vx^T \left( \alpha^2 \|\vx\| - \|\vx\| + \alpha \right)
\right) $$

with $$ \alpha = \frac{e^{\|\vx\|} - e^{-\|\vx\|}}{e^{\|\vx\|} + e^{-\|\vx\|}} $$

```python
def ddf(x):
    h = float(sqrt(x.T*x))
    alpha = float((exp(h) - exp(-h))/(exp(h) + exp(-h)))
    return (alpha * identity(x.size) - (x/h)*(x/h).T * (alpha - h + h*(alpha)**2))/h
```

```python
X = arange(-3.,3.,0.05)
Y = [ float(ddf(x)) for x in X ]

fig, axes = plt.subplots(figsize=(8,5))
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel(r'$\nabla^2 f(x)$')
axes.plot(X,Y)
```

For $n=2$, compute the first and second-order Taylor approximation around the points $\vx_0=\mathbf{0}$ and $\vx_0=(3,0)^T$.

Create a plot showing the error between the true function and the Taylor approximation as $\vx$ moves radially away from each $\vx_0$.
Use the radial direction of the form $\delta\vx = (\cos \phi, \sin \phi)^T$ for $\phi=\frac{2\pi k}{8}$ with $k \in \left\{0,1,...,7\right\}$. Specifically, for each $\delta\vx$, plot the error between the approximation and the true function at $\vx(\epsilon) = \vx_0 + \epsilon\delta\vx$ as $\epsilon$ increases from 0 to some sufficiently large value to see the general trend.

```python
n = 2
k = range(5)

eps = arange(0,1.,.01)
x0 = mat([0 ,0 ]).T
x1 = mat([3.,0 ]).T
```

##### First-order Taylor approximation
$$ f_1(\vx) = f(\vx_0) + \nabla f(\vx_0)^T(\vx-\vx_0)$$

```python
def f_1st(x, x0):
    return f(x0) + df(x0).T*(x-x0)
```

```python
err = np.zeros([len(k),eps.size])

fig, axes = plt.subplots(figsize=(12,8))

for i in range(len(k)):
    delta = mat([ cos(.25*pi*k[i]), sin(.25*pi*k[i]) ]).T
    for j in range(eps.size):
        x = x1 + eps[j]*delta
        err[i,j] = f(x) - f_1st(x, x1)
    axes.plot(eps,err[i,:],label='k='+str(k[i]))

axes.legend(loc='best')
axes.set_xlabel(r'$\epsilon$')
axes.set_ylabel(r'$f(x) - f_1(x)$')
axes.grid()
```

##### Second-order Taylor approximation
$$ f_2(\vx) = f(\vx_0) + \nabla f^T(\vx-\vx_0) + \frac{1}{2} (\vx-\vx_0)^T \nabla^2 f(\vx_0) (\vx-\vx_0)$$

```python
def f_2nd(x, x0):
    return f_1st(x,x0) + .5*(x-x0).T*ddf(x0)*(x-x0)
```

```python
err = np.zeros([len(k),eps.size])

fig, axes = plt.subplots(figsize=(12,8))

for i in range(len(k)):
    delta = mat([ cos(.25*pi*k[i]), sin(.25*pi*k[i]) ]).T
    for j in range(eps.size):
        x = x1 + eps[j]*delta
        err[i,j] = f(x) - f_2nd(x, x1)
    axes.plot(eps,err[i,:],label='k='+str(k[i]))

axes.legend(loc='best')
axes.set_xlabel(r'$\epsilon$')
axes.set_ylabel(r'$f(x)-f_2(x)$')
axes.grid()
```
