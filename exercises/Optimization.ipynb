## Problem 1: Implement binary logistic regression
$\newcommand{\vx}{\mathbf{x}}
\newcommand{\vw}{\mathbf{w}}$

Given a set of $N$ data points $\mathcal{D} = \left\{ (\vx_i, y_i) \right\}_{i=1}^N $
where $\vx_i \in \mathbb{R}^d$ is a vector of features and $y_i \in \{ -1, 1 \}$ is a binary label, we are searching for
a function $c(\vx) = \text{sign}(\vw^T\vx)$ that classifies points $\vx$ as either 1 or -1.

```python
X = loadtxt("classification_data/train_x.dat")
Y = loadtxt("classification_data/train_y.dat")
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))
ax.plot(X[Y==1][:,0], X[Y==1][:,1], 'ob', label="y =  1")
ax.plot(X[Y==-1][:,0], X[Y==-1][:,1], 'or', label="y = -1")
ax.set_xlim([-6.,6.]); ax.set_ylim([-4.,4.]);
ax.legend(loc='best'); ax.grid()
```

### Using logistic regression to find best parameters

From Homework 5 we know to represent the probability of whether a label $y$ is true as

$$ p(y|\vx,\vw) = \frac{1}{1+\exp(-y\vw^T\vx)} $$

for some parameter vector $\vw \in \mathbb{R}^d$. We also know for the binary case that
$p(1|\vx,\vw) = 1 - p(-1|\vx,\vw)$. The logistic regression method attempts to find the parameters $\vw$ that maximizes the likelihood of the above probability for the given data points. This can be formulated as an optimization problem 

\begin{align}
& \underset{\vw}{\text{min}} & & L(\vw) = - \log \prod_{i=1}^N p(y_i| \vx_i, \vw)
= \sum_{i=1}^N \log \left( 1 + \exp(-y_i\vw^T\vx_i) \right) \\
& \text{s.t.} & & \frac{1}{2} \|\vw\|^2
\end{align}

known as the neg-log-likelihood function subject to some regularization. We solve this by applying Newton's method

$$ \vw_{t+1} = \vw_t - \eta_t (\nabla^2 l(\vw_t))^{-1} \nabla l(\vw_t) $$

on

$$ l(\vw) = \sum_{i=1}^N \log \left( 1 + \exp(-y_i\vw^T\vx_i) \right) + \frac{\lambda}{2} \|\vw\|^2 $$

for a constant step size $\eta_t = 1$ and $\lambda = .7$, with gradient

$ \newcommand{\vX}{\mathbf{X}} \newcommand{\vy}{\mathbf{y}} \newcommand{\vp}{\mathbf{p}}$
$$\nabla l(\vw) = \sum_{i=1}^N -y_i(1-p_i)\vx_i + \lambda\vw$$

and Hessian

\begin{align}
\nabla^2 l(\vw) &= \sum_{i=1}^N p_i(1-p_i) \vx_i \vx_i^T + \lambda\mathbf{I} \\
&= \vX^T W \vX+\lambda\mathbf{I}
\end{align}

where $p_i = p(y_i|\vx_i,\vw)$ is the probabilty of a specific training point with the current parameters, $\vX = (\vx_1, \vx_2, ..., \vx_n)^T$ containing all $\vx_i$ as row vectors and $W$ as diagonal matrix with elements $w_{i,i}=p_i(1-p_i)$.


```python
def eval_objective(x,y,w,lam):
    return sum(log(1.+exp(-y*x.dot(w)))) + .5*lam*w.dot(w)

lam = .7
eta = 1.
w0 = random.rand(3)
l = [eval_objective(X,Y,w0,lam)]

gn = 1.; w = w0
while(gn > .0001):
    p = 1./(1.+exp(-Y*X.dot(w)))
    g = X.T.dot(-Y*(1.-p)) + lam*w
    H = X.T.dot(diag( p*(1.-p) )).dot(X) + lam*identity(len(w))
    w = w - eta * linalg.inv(H).dot(g)
    gn = linalg.norm(g)
    l.append(eval_objective(X,Y,w,lam))
```

### Applying the parameters on some test data

```python
X2 = loadtxt("classification_data/test_x.dat")
Y2 = loadtxt("classification_data/test_y.dat")
Y3 = sign(X2.dot(w))

xp = linspace(-6,6,12)
yp = -(w[0]*xp + w[2])/w[1]

fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))
ax.plot(X2[Y2==1][:,0], X2[Y2==1][:,1], 'ob', label="y =  1")
ax.plot(X2[Y2==-1][:,0], X2[Y2==-1][:,1], 'or', label="y = -1")
ax.plot(xp,yp,'g',label=r"$\mathbf{w}^T\mathbf{x} =$ 0")
ax.set_xlim([-6.,6.]); ax.set_ylim([-4.,4.]);
ax.legend(loc='best'); ax.grid()
```

```python
print "Score: {0}% \nIterations: {1}".format( (1.-len(X2[Y2!=Y3])/float(len(X2)))*100., len(l) )
```

```python
eta = arange(.1,.6,.1)
lg = [ [l[0]] for i in range(len(eta)) ]
for i in range(len(eta)):
    gn = 1.; w=w0
    while(gn > .0001):
        p = 1./(1.+exp(-Y*X.dot(w)))
        g = X.T.dot(-Y*(1.-p)) + lam*w
        w = w - eta[i] * 31./(30.+len(lg[i])) * g
        gn = linalg.norm(g)
        lg[i].append(eval_objective(X,Y,w,lam))
```

```python
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(16,8))
for i in range(len(eta)):
    ax.plot(lg[i],label=r"$\eta = $"+str(eta[i]))
ax.plot(l,label="Newton")
ax.legend(loc='best')
#ax.set_ylim([0,100])
ax.set_xlim([0,60])
ax.set_yscale('log')
ax.grid(True,which="both")
```
