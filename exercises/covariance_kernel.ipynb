```python
%matplotlib inline
from IPython.html.widgets import interactive, interact
from numpy import *
import matplotlib.pyplot as plt
```

```python
def plot(N, sigma):
    x = linspace(-5,5,N)
    X,Y = meshgrid(x,x)
    K = exp(-1./(2.*sigma**2)*(X-Y)**2)
    
    y = random.multivariate_normal(zeros(N),K,5)
    
    fig, ax = plt.subplots(ncols=2, figsize=(16,6))
    im = ax[1].imshow(K, interpolation='none', cmap=plt.get_cmap('Oranges'))
    cbar = plt.colorbar(im)
    for yi in y: ax[0].plot(x,yi,'x-')
    ax[0].set_xlim(-5.1,5.1)
    ax[0].set_ylim(-3.,3)
    ax[0].grid()
interactive(plot, N=(5,100,5), sigma=(.1, 5., .1))
```
